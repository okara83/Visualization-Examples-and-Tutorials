```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
plt.style.use('seaborn')
```

```python
gw = pd.read_csv("data/Global_Mean_Estimates_based_on_Land_and_Ocean_Data.csv",index_col=False)
gw.columns
```

```python
gw= gw.drop(columns=["Unnamed: 0"])
```

```python
gw
```

```python
gw.plot(x='Year',figsize=(20,5))
```

```python
ax=gw.plot(x='Year',y='Smoothed')
gw.plot.scatter(x='Year',y='Temp',c='Temp', colormap='inferno',figsize=(20,5),ax=ax)
```

```python
gw2=gw.drop(columns='Temp')
gw3=gw.drop(columns='Smoothed')
gw2
```

```python
gw2=gw2.set_index('Year')
sns.set(rc={'figure.figsize':(20,5)})
sns.heatmap(gw2.T)
gw2
```

```python
gw3=gw3.set_index('Year')
sns.set(rc={'figure.figsize':(20,5)})
sns.heatmap(gw3.T)
gw3
```

```python
gwdec = pd.read_csv("data/tempbydecade.csv")
gwdec=gwdec.set_index('Year')
gwdec
```

```python
sns.set(rc={'figure.figsize':(20,5)})
sns.heatmap(gwdec)
```
